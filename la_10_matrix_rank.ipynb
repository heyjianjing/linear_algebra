#### Rank of a matrix

For a matrix $A\in \mathbf{R}^{m \times n}$, the `rank` of $A$ is defined as `dimension` of its range


$$\text{rank}(A)=\dim R(A)$$

#### A matrix and its transpose have the same rank

(1) Assume column rank $\text{rank}(A)=r$

This means that every column in $A=\begin{bmatrix}a_1& \cdots& a_n\end{bmatrix}$ can be expressed as linear combination of $r$ basis vectors $\begin{bmatrix}q_1& \cdots& q_r\end{bmatrix}$ for $\mathbf{R}^r $

$$\begin{bmatrix}a_1& \cdots& a_n\end{bmatrix}_{m\times n}=QR=\begin{bmatrix}q_1& \cdots& q_r\end{bmatrix}_{m\times r}\begin{bmatrix}r_{11} & \cdots & r_{1n} \\ \vdots & \ddots & \vdots \\ r_{r1} & \cdots & r_{rn} \end{bmatrix}_{r\times n}$$

where

$$a_i = \sum_{j=1}^r q_jr_{ji}$$

On the other hand, we can also write the equation above using rows of $A$ and $Q$

$$\begin{bmatrix}\tilde{a}_1^T\\ \vdots\\ \tilde{a}_m^T\end{bmatrix}_{m\times n}=\begin{bmatrix}\tilde{q}_1^T\\ \vdots\\ \tilde{q}_m^T\end{bmatrix}_{m\times r}\begin{bmatrix}r_{11} & \cdots & r_{1n} \\ \vdots & \ddots & \vdots \\ r_{r1} & \cdots & r_{rn} \end{bmatrix}_{r\times n}$$

Therefore, each row in $A$ can also be expressed as the linear combination of $r$ rows in $R$

$$\tilde{a}_i^T = \tilde{q}_i^T\begin{bmatrix}r_{11} & \cdots & r_{1n} \\ \vdots & \ddots & \vdots \\ r_{r1} & \cdots & r_{rn} \end{bmatrix}$$

This means the row rank of $A$ is at most $r$, or $\text{rank}(A^T)\leq \text{rank}(A)=r$

(2) Assume row rank $\text{rank}(A^T)=r$

To get to column rank, we can go backwards

We can first show that each row of $A$ can be expressed as linear combination of $r$ rows in $R$ using row representation of $A$ and $Q$ as above

(note, here we only assume rows in $R$ form a basis for $\mathbf{R}^r$)

Then, we rewrite the equation using column representation of $A$ and $Q$, and show that each column of $A$ can be expressed as linear combination of $r$ columns in $Q$

As a result, the column rank of $A$ is at most $r$, or $\text{rank}(A)\leq\text{rank}(A^T)=r$

Therefore

$$\text{rank}(A)=\text{rank}(A^T)$$

#### Full rank matrices

* for square matrices, full rank means non-singular
* for tall matrices, full rank means `columns` are independent
* for fat matrices, full rank means `rows` are independent
