#### Eigenvalues and eigenvectors

$\lambda\in C$ is an `eigenvalue` of $A\in C^{n \times n}$ if

* there exists nonzero $v\in C^n$ such that
$$(\lambda I-A)v=0, \text{  i.e.,}\,\, Av=\lambda v$$
any such vector $v$ is an (right) `eigenvector` of $A$

* there exists nonzero $w\in C^n$ such that
$$w^T(\lambda I-A)=0, \text{  i.e.,}\,\, w^TA=\lambda w^T$$
any such vector $w$ is a `left eigenvector` of $A$

Even if $A$ is `real`, eigenvalue and eigenvector can be `complex`

#### `Properties`

(1) $A$ and $A^T$ have the `same eigenvalues`

* eigenvalues are the roots of $\det(sI-A)$
* from properties of determinant, we have $\det(sI-A)=\det(sI-A)^T=\det(sI-A^T)$

So, eigenvalues of $A$ and $A^T$ are the same

(2) $A$ is `invertible` $\Longleftrightarrow$ $A$ does `not` have zero eigenvalue

* If 0 is an eigenvalue of $A$, then $\det(sI-A)=0$ when $s=0$. Therefore, $\det(-A)=\det(A)=0$, and $A$ is `not` invertible. So, if $A$ is invertible, then, eigenvalue cannot be zero
* If $A$ is `not` invertible, then, $\det(-A)=\det(A)=0$. This means that 0 is an eigenvalue of $A$. So, if $A$ does `not` have zero eigenvalue, then, $A$ must be invertible


(3) If $A$ has eigenvalues $\lambda_i$'s and $A$ is invertible, then $A^{-1}$ has eigenvalues $\lambda_i^{-1}$'s

Since $A$ is invertible, we know 0 is `not` an eigenvalue of $A$. For a pair of $\lambda_i$ and $v_i$, we have

$$Av_i=\lambda_iv_i$$

Multiply both side by $A^{-1}$ and $\lambda_i^{-1}$, we have

$$\lambda_i^{-1}v_i=A^{-1}v_i$$

So the eigenvalues of the inverse are the `inverse of eigenvalues`

(4) Eigenvalues of $A$ and $T^{-1}AT$ are the same

$$\det(sI-A)=\det(I(sI-A))=\det(T^{-1}T(sI-A))$$

From property of determinant ($\det (AB)=\det(A)\det(B)$) we have

$$\det(T^{-1}T(sI-A))=\det(T^{-1}(sI-A)T)$$

and (since $T^{-1}sIT=sI$)

$$\det(T^{-1}(sI-A)T)=\det(sI-T^{-1}AT)$$

Therefore, the eigenvalues of $A$ and $T^{-1}AT$ are the same


#### `Spectrum` of a matrix is the set of its eigenvalues

#### `Markov chain`

$z(t)\in\{1, \cdots, n\}$ (indicating $z(t)$ can be in $n$ discrete states) is a random sequence over $t$ with the following transition

$$p(z(t+1)=i|z(t)=j)=P_{ij}$$

where $P\in \mathbf{R}^{n\times n}$ is the `matrix of transition probabilities`

If we write

$$p(t)=\begin{bmatrix}p_1(t)\\p_2(t)\\...\\p_n(t)\end{bmatrix}=\begin{bmatrix}p(z(t)=1)\\p(z(t)=2)\\...\\p(z(t)=n)\end{bmatrix}$$

We have

$$\boxed{p(t+1)=Pp(t)}$$

#### Connection to `eigenvector`

Apparently, to be a valid Markov chain, a few constraints needs to be satisfied
* for $p(t)$, the sum of all $p_i(t)=p(z(t)=i)$ must be `one`, that is, $\sum_{i=1}^np_i(t)=1$
* each `column` in matrix $P$ must have a sum of `one`, that is, $\sum_{i=1}^nP_{ij}=1$

Based on this, we can write the following to sum each columns of $P$

$$\begin{bmatrix}1 & 1 & \cdots & 1\end{bmatrix}P=\begin{bmatrix}1 & 1 & \cdots & 1\end{bmatrix}$$

This means that $\begin{bmatrix}1 & 1 & \cdots & 1\end{bmatrix}$ is a `left eigenvector` of $P$ with `eigenvalue` of 1

#### What is `right eigenvector` of $P$?

There is a `right eigenvector` $v\neq0$ such that

$$Pv=1\cdot v$$

This means, this right eigenvector is an `equilibrium distribution`, or `steady-state distribution` of the Markov chain
