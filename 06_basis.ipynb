#### A completely non rigorous definition of a `vector space`

A `collection` of vectors that can be added together and scaled by scalars, `while still staying within the collection`. These vectors follow certain rules, such as:

* Addition: You can add any two vectors together to get another vector in the same space.
* Scalar Multiplication: You can multiply any vector by a scalar (a number) to get another vector in the same space.

These operations follow specific rules (like associativity, commutativity for addition, etc.) that make the vector space a structured and predictable environment for working with vectors.

For example, n-dimensional space $\mathbf{R}^n$ is a vector space

#### `Subspace`

Intuitively, a `subspace of a vector space` is a `subset` of a vector space

The following three are all `subspaces` of $\mathbf{R}^n$

$$\begin{align}
V_1&=\mathbf{R}^n \\
V_2&=\{0\}, 0 \in \mathbf{R}^n \\
V_3&=\text{span}(v_1, v_2, \cdots, v_k), v_i \in \mathbf{R}^n
\end{align}$$

#### `Independent` set of vectors

A set of vectors $v_1, v_2, \cdots, v_k$ is `independent` if

$$\boxed{\alpha_1v_1+\alpha_2v_2+\cdots +\alpha_kv_k=0 \Rightarrow \alpha_1=\alpha_2=\cdots = \alpha_k = 0}$$

We then can have the following two

First

$$\alpha_1v_1+\alpha_2v_2+\cdots +\alpha_kv_k=\beta_1v_1+\beta_2v_2+\cdots +\beta_kv_k \Rightarrow \alpha_i=\beta_i, i=1, 2,\cdots, k$$

Second, there does `not` exist set of coefficients $\alpha_i$'s, such that

$$v_j=\alpha_1v_1+\cdots+\alpha_{j-1}v_{j-1}+\alpha_{j+1}v_{j+1}+\cdots+\alpha_kv_k$$

That is, no $v_j$ can be expressed as the linear combination of the remaining basis vectors (can be seen via contradiction)

#### `Basis`

We have mentioned basis vectors $e_i$'s and the transformed version $e'_i$'s previously

More formally, $v_1, v_2, \cdots, v_k$ is a set of `basis vectors` for vector space V if

* $V=\text{span}(v_1, v_2, \cdots, v_k)$
* $v_1, v_2, \cdots, v_k$ forms an `independent` set

For every $v\in V$, it can be `uniquely` expressed using basis vectors

$$v=\alpha_1v_1+\alpha_2v_2+\cdots +\alpha_kv_k$$

The `simpliest basis` for $\mathbf{R}^n$ is $e_1, e_2, \cdots, e_n$, which are also `orthonormal`, that is

$$e_i\cdot e_j = \left\{\begin{array}{rcl}0 &i\neq j \\1 &i=j \end{array}\right.$$

For a given vector space $V$, the `number of vectors` in any basis is the same

Number of vectors in any basis is the `dimension` of the corresponding vector space $\dim V$
