#### `Norm` of a vector

For a vector $x\in \mathbf{R}^n$, its `Euclidean norm` is defined as


$$\|x\|=\sqrt{x_1^2+x_2^2+\cdots+x_n^2}=\sqrt{x^Tx}$$

In general, for $1\leq p\leq +\infty$, the $l_p$ norm is defined as

$$\|x\|_p=\left(\sum_{i=1}^n|x_i|^p\right)^{\frac{1}{p}}$$

So the Euclidean norm is $l_2$ norm, where subscript 2 is often omitted

Euclidean norm (and other $l_p$ norms) satisfies the following

* $\|ax\|=|a|$$\|x\|$
* $\|x+y\| \leq \|x\|+\|y\|$
* $\|x\|=0 \Longleftrightarrow x=0$

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')
np.set_printoptions(formatter={'float': '{: 0.2e}'.format})
```

```python
def plot_norm(p):
    x1 = np.linspace(-5, 5, 100)
    x2 = np.linspace(-5, 5, 100)
    x1, x2 = np.meshgrid(x1, x2)

    p=p
    l2_norm = (np.abs(x1)**p + np.abs(x2)**p)**(1/p)

    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x1, x2, l2_norm, cmap='viridis')

    ax.set_title(f'L_{p} Norm of (x1, x2)')
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.set_zlabel(f'L_{p} Norm')
    plt.tight_layout()
    plt.show()
```

```python
plot_norm(2)
```

```python
plot_norm(100)
```

We can picture, the `infinity` norm is $$\|x\|_{\infty}=\max_i (|x_i|)$$

#### `CauchySchwarz` inequality

For any $x$, $y\in \mathbf{R}^n$, we have

$$\boxed{|x^Ty|\leq\|x\|\|y\|}$$

One of the `proofs`

$$0\leq\frac{1}{2}\left\|\frac{x}{\|x\|}\pm \frac{y}{\|y\|}\right\|^2=1\pm \frac{x^Ty}{\|x\|\|y\|}$$

And, we are done

$$\pm \,x^Ty\leq\|x\|\|y\|$$


#### `Inequality` between $l_1$ and $l_2$ norm

With (squared) Cauchy-Schwarz, we can write the following

$$\left(\sum|x_iy_i|\right)^2\leq \left(\sum x_i^2\right) \left(\sum y_i^2\right)$$

Let $y_i=1$, we have

$$\left(\sum|x_i|\right)^2\leq \left(\sum x_i^2\right) \left(\sum 1^2\right)$$

or

$$\|x\|_1^2\leq n \|x\|_2^2$$

and

$$\boxed{\|x\|_1\leq \sqrt{n} \|x\|_2}$$
