#### `Nullspace` of a matrix

For a matrix $A\in \mathbf{R}^{m \times n}$, its `nullspace` is defined as

$$N(A)=\{x\in \mathbf{R}^n |Ax=0\}$$

* $N(A)$ is set of vectors `mapped to zero` by transformation $y=Ax$
* any vector $x$ in $N(A)$ is `orthogonal to rows` of `A` (recall `2nd` interpretation of $y=Ax$ as inner product between $x$ and rows of $A$)
* if $y=Ax$ and $z\in N(A)$, then $y=A(x+z)$, that is $x$ and $x+z$ are `indistinguishable` from $y$
* $0\in \mathbf{R}^n$ is always in $N(A)$

#### `Zero` nullspace, $N(A)=\{0\}$

$A$ is called `one-to-one` if the only element in the nullspace is $0\in \mathbf{R}^n$ $\Longleftrightarrow$

* $x$ can always be `uniquely` determined by $y=Ax$
* mapping from $x$ to $y=Ax$ is `one-to-one`, that is, different $x$ $\Rightarrow$ different $y$
* `columns` of $A$ are `independent` set. If they are not independent, then there exists a column $a_j$ that can be expressed as linear combination of the rest columns $a_j=\beta_1a_1\cdots+\beta_{j-1}a_{j-1}+\beta_{j+1}a_{j+1}+\cdots +\beta_na_n$. If we take $x=[\beta_1,\cdots,\beta_{j-1}, -1, \beta_{j+1}, \cdots, \beta_n]^T$, we can see that $x\in N(A)$, so contradiction
* as a result of being independent set, `columns` of $A$ also form a `basis` for vector space that they span, which is a `subspace` of $\mathbf{R}^m$
* $A$ has `left inverse` $B\in \mathbf{R}^{n \times m}$, such that $BA=I$, where $I$ is the `identity`matrix. As $By=BAx=x$, $B$ can be considered a `perfect decoder` to retrieve $x$ from $y$ as a `linear combination` of columns of $B$
* $\det (A^TA)\neq 0$

#### On `left inverse`

(1) show $A$ has left inverse B $\Rightarrow$ $N(A)={0}$

Let $Ax=0$ for some $x$ (this $x$ must exist as $0\in \mathbf{R}^n$ always in nullspace), then

$$0=B(Ax)=(BA)x=Ix=x$$

This means, the only `linear combination` of columns of $A$ to satisfy $Ax=0$ is for $x$ to be 0. Therefore, $N(A)=0$

Also, `linear independent` set of `columns` constrain matrix $A$ to be `tall` or `square`

(2) show $N(A)={0}$  $\Rightarrow$ $A$ has left inverse B

Let $x$ be a vector that satisfies

$$(A^TA)x=0$$

Multiply on the left by $x^T$

$$x^TA^TAx=\|Ax\|^2=0$$

This implies $Ax=0$. Since columns of $A$ are independent due to $N(A)=0$, we get that the only solution for $(A^TA)x=0$ is $x=0$

Therefore, $A^TA$ is `invertible`

Next, we simply construct

$$B=(A^TA)^{-1}A^T$$

and verify that

$$BA=(A^TA)^{-1}A^TA=I$$

That is, $B$ is a left inverse of $A$

#### On the `determinant` of $A^TA$

(1) show $N(A)\neq \{0\} \Rightarrow \det(A^TA)=0$

Let $x\in N(A)$ and $x\neq 0$, then

$$Ax=0$$

Left multiply $A^T$ on the both side

$$A^TAx=0$$

Since $x\neq 0$, we have $\det(A^TA)=0$

So, $\boxed{\det(A^TA)\neq 0 \Rightarrow N(A)={0}}$

(2) show $\det(A^TA)=0  \Rightarrow N(A)\neq \{0\}$

As $\det(A^TA)=0$, then for $x\neq 0$ (`geometric` interpretation of determinant)

$$(A^TA)x=0$$

Left multiply $x^T$ on both side

$$x^T(A^TA)x=\|Ax\|^2=0$$

Therefore, $x \in N(A)$

$$Ax=0$$

Since $x\neq 0$, we have $N(A)\neq \{0\}$

So, $\boxed{ N(A)=\{0\}\Rightarrow \det(A^TA)\neq 0}$
