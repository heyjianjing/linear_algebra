#### Transformation of `basis vector`

A matrix `A` can be considered a system that `transforms` an input `vector` $v_i$ to an output `vector` $v_o$ through `transforming basis vectors`

For example, for a 2D matrix `A`

$$A=\begin{bmatrix} 3 & 2 \\ -2 & 1\end{bmatrix}$$

Its `1st column`

$$\begin{bmatrix}3 \\-2 \end{bmatrix}$$

indicates where `basis vector` $e_1=\begin{bmatrix}1\\0\end{bmatrix}$ lands after the `transformation` induced by A, that is

$$\boxed{e'_1=A\cdot e_1=\begin{bmatrix} 3 & 2 \\ -2 & 1\end{bmatrix}\begin{bmatrix}1\\0\end{bmatrix}=\begin{bmatrix}3 \\-2 \end{bmatrix}}$$

And its `2nd column`

$$\begin{bmatrix}2 \\1 \end{bmatrix}$$

indicates where `basis vector` $e_2=\begin{bmatrix}0\\1\end{bmatrix}$ lands after the `transformation` induced by A, that is

$$\boxed{e'_2=A\cdot e_2=\begin{bmatrix} 3 & 2 \\ -2 & 1\end{bmatrix}\begin{bmatrix}0\\1\end{bmatrix}=\begin{bmatrix}2 \\1 \end{bmatrix}}$$

`Note`, both $e'_1$ and $e'_2$ are expressed in `original` $e_1,\,e_2$ coordinate

#### `Composition` of input and output of transformation

For any `input vector` $v_i$ that can be expressed as

$$v_i=\begin{bmatrix}a\\b\end{bmatrix}=a\cdot \begin{bmatrix}1\\0\end{bmatrix} + b \cdot \begin{bmatrix}0\\1\end{bmatrix}=a\cdot e_1 +b\cdot e_2$$

where $a$ and $b$ are `scalar`, the `output vector` $v_o$ after transformation is expressed as

$$v_o=Av_i=a\cdot A\cdot e_1+b\cdot A\cdot e_2=a\cdot\begin{bmatrix}3\\-2\end{bmatrix}+b\cdot\begin{bmatrix}2\\1\end{bmatrix}=a\cdot e'_1+b\cdot e'_2$$

This essentially says that the vector $v_o$ after transformation is simply the `weighted average` of `transformed` basis vectors $\boxed{e'_1,\, e'_2}$, with the weights being the same as the vector $v_i$ based on basis vectors $e_1,\, e_2$

Here, `output` vector $v_o$ is still expressed in `original` $e_1, e_2$ coordinate, `NOT in transformed` $e'_1, e'_2$ coordinate

#### `Dual interpretation` of $v_o=Av_i$

##### `1st` interpretation as `transformation`

When $v_i$ is expressed in `original` $e_i$ coordinate
  
Then, $Av_i$ moves $v_i$ along with the transformation induced by A

The `transformed`, new vector $v_o$ is still expressed in `original` $e_i$ coordinate

i.e., $v_i$ and $v_o$ are two `different` vectors expressed in the same `original` $e_i$ coordinate

This is what we have seen so far

##### `2nd` interpretation as `translation`

Now let $v_i$ be a vector expressed in `transformed` $e'_i$ coordinate

Then, $Av_i$ `translates` how $v_i$ is expressed in `transformed` coordinate back to how $v_i$ should be expressed in `original` $e_i$ coordinate

The `transformation` between $e'_i$ coordinate and $e_i$ coordinate is defined by `A`

i.e., $v_o$ and $v_i$ represent the `same` vector in `transformed` $e'_i$ coordinate and `original` $e_i$ coordinate, respectively

This is like `change of perspective` of the same thing

`For example`, if $e'_1=\begin{bmatrix}2\\1\end{bmatrix}$ and $e'_2=\begin{bmatrix}-1\\1\end{bmatrix}$, then we know the transformation is induced by matrix

$$A=\begin{bmatrix}2 & -1\\1& 1\end{bmatrix}$$

If there is a vector $v_i$ expressed as $\begin{bmatrix}-1\\2\end{bmatrix}$ in $e'_i$ coordinate, then it is expressed as

$$Av_i=\begin{bmatrix}2 & -1\\1& 1\end{bmatrix}\begin{bmatrix}-1\\2\end{bmatrix}=\begin{bmatrix}-4\\1\end{bmatrix}$$

in $e_i$ coordinate

#### `Change of basis`

The `2nd` interpretation is important when we want to manipulate a vector in $e'_i$ coordinate, but the manipulation is much easier in $e_i$ coordinate

In this case, we can use `A` to first `translate` the vector into $e_i$ coordinate, do the manipulation, then `translate` the manipulated vector back to $e'_i$ coordinate

For example, for

$$A=\begin{bmatrix}2 & -1\\1& 1\end{bmatrix}$$

If there is a vector $v$ in `transformed` $e'_i$ coordinate and we want to rotate it by 90Â° CCW and we want to know how the `rotated vector` is expressed in `transformed` $e'_i$ coordinate, we perform the following

`First`, we translate $v$ from `transformed` $e'_i$ coordinate to express in `original` $e_i$ coordinate (`2nd` interpretation, same vector, different coordinate)

$$Av=\begin{bmatrix}2 & -1\\1& 1\end{bmatrix}v$$

`Then`, with expression in `original` $e_i$ coordinate, we apply rotation matrix to describe how the `rotated vector` is expressed in `original` $e_i$ coordinate (`1st` interpretation, different vector, same coordinate)

$$RAv=\begin{bmatrix}0 & -1\\1& 0\end{bmatrix}\begin{bmatrix}2 & -1\\1& 1\end{bmatrix}v$$

`Finally`, we `translate` rotated vector back to `transformed` $e'_i$ coordinate (`2nd` interpretation, same vector, different coordinate)

$$A^{-1}RAv=\begin{bmatrix}2 & -1\\1& 1\end{bmatrix}^{-1}\begin{bmatrix}0 & -1\\1& 0\end{bmatrix}\begin{bmatrix}2 & -1\\1& 1\end{bmatrix}v$$
