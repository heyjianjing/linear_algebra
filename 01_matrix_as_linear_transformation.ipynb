#### Transformation of `basis vector`

A matrix `A` can be considered a system that `transforms` an input `vector` $v_i$ to an output `vector` $v_o$ through $\boxed{v_o=Av_i}$

For example, for a 2D matrix `A`

$$A=\begin{bmatrix} 3 & 2 \\ -2 & 1\end{bmatrix}$$

Its `1st column`

$$\begin{bmatrix}3 \\-2 \end{bmatrix}$$

indicates where `basis vector` $e_1=\begin{bmatrix}1\\0\end{bmatrix}$ lands after the `transformation` induced by A, that is

$$\boxed{A\cdot e_1=\begin{bmatrix} 3 & 2 \\ -2 & 1\end{bmatrix}\begin{bmatrix}1\\0\end{bmatrix}=\begin{bmatrix}3 \\-2 \end{bmatrix}}$$

And its `2nd column`

$$\begin{bmatrix}2 \\1 \end{bmatrix}$$

indicates where `basis vector` $e_2=\begin{bmatrix}0\\1\end{bmatrix}$ lands after the `transformation` induced by A, that is

$$\boxed{A\cdot e_2=\begin{bmatrix} 3 & 2 \\ -2 & 1\end{bmatrix}\begin{bmatrix}0\\1\end{bmatrix}=\begin{bmatrix}2 \\1 \end{bmatrix}}$$

#### `Composition` of input and output of transformation

For any `input vector` $v_i$ that can be expressed as

$$v_i=\begin{bmatrix}a\\b\end{bmatrix}=a\cdot e_1 +b\cdot e_2=a\cdot \begin{bmatrix}1\\0\end{bmatrix} + b \cdot \begin{bmatrix}0\\1\end{bmatrix}$$

where $a$ and $b$ are `scalar`, the `output vector` $v_o$ after transformation is expressed as

$$v_o=Av_i=a\cdot A\cdot e_1+b\cdot A\cdot e_2=a\cdot\begin{bmatrix}3\\-2\end{bmatrix}+b\cdot\begin{bmatrix}2\\1\end{bmatrix}$$

This essentially says that the vector $v_o$ after transformation is simply the `weighted average` of `transformed` basis vectors $e_1$ and $e_2$, with the weights being the same as the original vector $v_i$

Here, the `output` vector $v_0$ is still expressed in `coordinate system before transformation`, that is, the `original` $e_1, e_2$ coordinate system
