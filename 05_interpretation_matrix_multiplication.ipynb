#### `Interpretation` of $y=Ax$

$y\in \mathbf{R}^m,\,A\in \mathbf{R}^{m\times n},\, x\in \mathbf{R}^n$

##### `1st` interpretation, as `mixture of columns` of $A$

$$A=\begin{bmatrix}a_1& a_2 & \cdots & a_n\end{bmatrix}$$

where each column $a_i\in\mathbf{R}^m$

Then, for

$$x=\begin{bmatrix}x_1\\x_2\\...\\x_n\end{bmatrix}$$

we have

$$\boxed{y=\sum_{i=1}^n x_ia_i}$$

##### `2nd` interpretation, as `inner product with rows` of $A$

$$A=\begin{bmatrix}\tilde{a_1}^T\\\tilde{a_2}^T\\...\\\tilde{a_m}^T\end{bmatrix}$$

where each row $\tilde{a_i}^T\in\mathbf{R}^n$

Then, we have

$$\boxed{y=\begin{bmatrix}\tilde{a_1}^T\cdot x\\\tilde{a_2}^T\cdot x\\...\\\tilde{a_m}^T\cdot x\end{bmatrix}}$$

#### `Interpretation` of $C=AB$

$C\in \mathbf{R}^{m\times p},\,A\in \mathbf{R}^{m\times n},\, B\in \mathbf{R}^{n\times p}$

##### `1st` interpretation, `A acting on columns of B`

$$\boxed{C=A\begin{bmatrix}b_1& b_2 & \cdots & b_p\end{bmatrix}}$$

where each column $b_i\in\mathbf{R}^n$

That is, `each column in C is some linear combination of columns in A`

##### `2nd` interpretation, `B acting on rows of A`

$$\boxed{C=\begin{bmatrix}\tilde{a_1}^TB\\\tilde{a_2}^TB\\...\\\tilde{a_m}^TB\end{bmatrix}}$$

where each row $\tilde{a_i}^T\in\mathbf{R}^n$

That is, `each row in C is some linear combination of rows in B`

##### `3rd` interpretation, `sum of outer products, or rank-1 matrices`

Write

$$A=\begin{bmatrix}a_1& a_2 & \cdots & a_n\end{bmatrix}$$

and

$$B=\begin{bmatrix}\tilde{b_1}^T\\\tilde{b_2}^T\\...\\\tilde{b_n}^T\end{bmatrix}$$

where $a_i\in \mathbf{R}^m$ and $\tilde{b_i}^T \in \mathbf{R}^p$

Then

$$\boxed{C=\sum_{i=i}^na_i\tilde{b_i}^T}$$

Each $a_i\tilde{b_i}^T\in \mathbf{R}^{m\times p}$ is a `low-rank approximation` of C
