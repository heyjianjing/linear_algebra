#### `Euclidean norm` of a vector

For a vector $x\in \mathbf{R}^n$, its `Euclidean norm` is defined as


$$\|x\|=\sqrt{x_1^2+x_2^2+\cdots+x_n^2}=\sqrt{x^Tx}$$

It measures length of vector from origin

* $\|ax\|=|a|$$\|x\|$

* $\|x+y\| \leq \|x\|+\|y\|$
* $\|x\|=0 \Longleftrightarrow x=0$
