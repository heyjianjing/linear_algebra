#### Definition of `determinant`

Determinant of a matrix A is the `signed scaling factor` of unit square, cube, etc. defined by `basis vectors` $e_i$'s, before and after transformation induced by A. The `sign` of determinant is determined based on `right hand rule`

For example, in 2D, determinant of matrix A is the `area of parallelogram` defined by transformed $e_1$ and $e_2$ after transformation induced by A, namely $e'_1$ and $e'_2$

If after transformation $e'_1$ and $e'_2$ no longer follow right hand rule, then, determinant is `negative`

#### What does this have to do with `cross-product`?

The cross-product of two vectors

$$p=v \times w$$ is a `vector` perpendicular to both per `right hand rule` and a `length` equals the `area of parallelogram` defined by $v$ and $w$

With this, the effect of `dot product` between $p$ and a vector $x$ is `equivalent` to the `volume of parallelepiped` defined by $x$, $v$, and $w$

This is because

* $p\cdot x$ finds the `scaled projection length` of $x$ onto $p$, where the actual `projection length` is the length of $x$ in the direction perpendicular to both $v$ and $w$, that is, the `height` of the parallelepiped
* the `scaling factor` is the length of $p$, which is `area` of parallelogram

As a result, $p \cdot x$ is equivalent to the determinant of matrix whose columns are $x$, $v$, and $w$

$$\begin{bmatrix}p_1\\p_2\\p_3\end{bmatrix}\cdot \begin{bmatrix}x_1\\x_2\\x_3\end{bmatrix}=\det\left(\begin{bmatrix}x_1& v_1 & w_1\\x_2& v_2 & w_2\\x_3& v_3 & w_3\end{bmatrix}\right)$$

#### Intuition on why A is `not` invertible if $\det (A)=0$

This is because with $\det (A)=0$, an input vector is `crushed to zero` after transformation (or the whole vector space is crushed to zero). As a result, information is lost after transformation, that is, $x$ can not be `recovered` from $Ax=0$
