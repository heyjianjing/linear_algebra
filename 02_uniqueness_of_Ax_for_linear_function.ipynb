#### For a `linear` function $f: \mathbf{R}^n\rightarrow \mathbf{R}^m$, there is a `matrix` $A\in \mathbf{R}^{m\times n}$ such that $f(x)=Ax$

For

$$x=\begin{bmatrix}x_1\\x_2\\...\\x_n\end{bmatrix}$$

We can write $x$ using `basis` vectors $e_n$'s

$$x=x_1e_1+x_2e_2+\cdots + x_ne_n$$

where $x_n$'s are `scalar`. Using `linearity` of $f(\cdot)$

$$\boxed{f(ax+bx')=af(x)+bf(x')}$$

we have

$$f(x)=x_1f(e_1)+x_2f(e_2)+\cdots +x_nf(e_n)$$

or in `matrix` form

$$f(x)=\begin{bmatrix}f(e_1) & f(e_2) & \cdots & f(e_n)\end{bmatrix}\cdot\begin{bmatrix}x_1\\x_2\\...\\x_n\end{bmatrix}$$

As a result, if take

$$\boxed{A=\begin{bmatrix}f(e_1) & f(e_2) & \cdots & f(e_n)\end{bmatrix}\in \mathbf{R}^{m\times n}}$$

We get what we need

$$f(x)=Ax$$

This means `any linear function` can be expressed as $Ax$

#### A is `unique`

Suppose we have $A'\in \mathbf{R}^{m\times n}$, different from A, such that

$$f(x)=A'x$$

Then, the following must stand for all $x\in \mathbf{R}^n$

$$(A-A')x=0$$

If we select $x$ as one of the `basis` vector $e_i$, then

$$(A-A')e_i=0$$

This means that the `ith` column of `A` and `A'` must be the same

So plugging in $e_i$, for $i=1, 2,\cdots , n$, we can get that `all columns` of $A-A'$ are zero. Therefore, $A=A'$, and $A$ is `unique`

#### $y=Ax$ as the `only expression` of linear function

From earlier, we see that if $f(\cdot)$ is linear, then $f(x)=Ax$

Since showing the linearity of $g(\cdot)=Ax$ is trivial

We can conclude that $y=Ax$ is the `only way` of expressing linear function
